```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data
```

```python
# Compare population in the US and China
us = data[data.country == 'United States'] # Fetching only the US
us
```

```python
# Only China
china = data[data.country == 'China']
china
```

```python
# Overall graph for the US and China
plt.plot(us.year, us.population / 10 ** 6) # 10 ** 6 - разделение на миллионы 
plt.plot(china.year, china.population / 10 ** 6)

plt.legend(['US', 'China'])
plt.xlabel('Population')
plt.ylabel('Year')

plt.show()
```

```python
# Percentage relative (то же самое только в процентом соотношении)
plt.plot(us.year, us.population / us.population.iloc[0] * 100) # рост в процентном соотношении, относительно первого знач.
plt.plot(china.year, china.population / china.population.iloc[0] * 100)

plt.legend(['US', 'China'])
plt.xlabel('Population')
plt.ylabel('Year')

plt.show()
```
