```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('countries_covid_agg.csv')
uk = df[df['Country'].str.contains('United Kingdom')].reset_index(drop=True)
fr = df[df['Country'] == 'France'].reset_index(drop=True)
ge = df[df['Country'].str.contains('Germany')].reset_index(drop=True)
ge.tail(3)
```

```python
plt.figure(figsize=(13, 7))

plt.plot(uk['Date'][:251], uk['Confirmed'][:251], color='red', label='UK Total')
plt.plot(uk['Date'][250:], uk['Confirmed'][250:], 'r--')
plt.plot(fr['Date'][:266], fr['Confirmed'][:266], color='green', label='FRA Total')
plt.plot(fr['Date'][265:], fr['Confirmed'][265:], 'g--')
plt.plot(ge['Date'][:276], ge['Confirmed'][:276], color='blue', label='GER Total')
plt.plot(ge['Date'][275:366], ge['Confirmed'][275:366], 'b--')
plt.plot(ge['Date'][365:], ge['Confirmed'][365:], color='blue')

plt.title('COVID-19', loc='right')
plt.xlabel('Date', fontsize=20)
plt.xticks(uk['Date'][::30], rotation=45)
plt.ylabel('Cases (Million)')
plt.yticks(np.arange(0, 6000000, 500000))
plt.legend()
plt.show()
```
