```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Line Graph

```python
gas = pd.read_csv('gas_prices.csv')
gas.head(3)
```

```python
plt.figure(figsize=(13, 7))

# plt.plot(gas.Year, gas.USA, label='USA')  # USA
# plt.plot(gas['Year'], gas.Canada, label='Canada')  # Canada
# plt.plot(gas['Year'], gas['South Korea'], ".-", label = 'S. Korea', )  # S. Korea

# OR
for i in gas:
    if i != 'Year':
        plt.plot(gas.Year, gas[i], '.-', label = i)
        
# Cosmetics
plt.title('Gas Price Over Time', fontdict={'fontweight':'bold', 'fontsize': 18})
plt.xticks(gas.Year[::4].tolist() + [2020])  # Year range x-axis
plt.legend()
plt.xlabel('YEAR')
plt.ylabel('US Dollars')

plt.savefig('gas_price.png', dpi=300)

plt.show()
```
