```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_gas_price.csv')
df.head(10)
```

```python
plt.figure(figsize=(9, 7))

for i in df:
    if i != 'Year' and i != 'USA':
        plt.plot(df['Year'], df[i], '.-', label=i)
    elif i == 'USA':
        plt.plot(df['Year'], df[i], ':', label=i)


plt.title('Gas Price Comparison', fontdict={'fontweight':'bold', 'fontsize':20}, loc='right')
plt.xlabel('YEAR', fontdict={'fontsize':16})
plt.ylabel('$ per gallon', fontdict={'fontsize':16})
plt.xticks(df['Year'][::2].to_list() + [2012], rotation='60')
plt.legend()
# plt.savefig('gas_price.png', dpi=300)
plt.show()
```
