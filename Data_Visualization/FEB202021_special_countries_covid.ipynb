```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('countries_covid_agg.csv')
df.columns
# Data taken from datahub.io
```

# Spreading

```python
# Pivoting
df2 = df.pivot(index='Date', columns='Country', values='Confirmed').reset_index()
df2 = df2.loc[df2['Date'] >= '2020-03-01'].reset_index(drop=True)
```

```python
# parameters
ctr = ['Italy', 'Sweden', 'Korea, South', 'United Kingdom', 'France']
```

```python
# PLOT
plt.figure(figsize=(11, 6))

for i in df2:
    if i != 'Date' and i in ctr:
        plt.plot(df2['Date'], df2[i], label=i)

# Legend
plt.title('CONFIRMED CASES')
plt.xlabel('DATE')
plt.ylabel('Million Cases')
plt.xticks(df2['Date'][::30], rotation=60)
plt.legend()        
plt.show()
```

# Death Rate

```python
df3 = df.pivot(index='Date', columns='Country', values='Deaths').reset_index()
df3 = df3.loc[df3['Date'] >= '2020-03-01'].reset_index(drop=True)
```

```python
for i in df3:
    if i != 'Date' and i in ctr:
        plt.plot(df3['Date'], df3[i], '', label=i)

        
plt.xticks(df3['Date'][::30], rotation=60)
plt.legend()
plt.show()
```
